```python
import numpy as np
from datasets import load_dataset
fmnist_dataset = load_dataset("fashion_mnist")
fmnist_dataset
train = fmnist_dataset["train"]
test = fmnist_dataset["test"]
```

```python
class LogisticRegression:
  def __init__(self, lr = 0.01, n_iters = 1000):
    self.lr = lr
    self.n_iters = n_iters
    self.weights = None
    self.bias = None
  def sigmoid(self,z):
    return 1/(1+np.exp(-z))
  def fit(self, X, y):
    n_samples, n_features = X.shape
    self.weights = np.zeros(n_features)
    self.bias = 0
    for i in range(self.n_iters):
      z = np.dot(X, self.weights) + self.bias
      y_pred = self.sigmoid(z)
      dw = (1/n_samples) * np.dot(X.T, (y_pred - y))
      db = (1/n_samples) * np.sum(y_pred - y)
      self.weights -= self.lr * dw
      self.bias -= self.lr * db
  def predict_proba(self, X):
    z = np.dot(X, self.weights) + self.bias
    return self.sigmoid(z)
  def predict(self,X):
    probs = self.predict_proba(X)
    return (probs >= 0.5).astype(int)
```

```python
X_train = np.array([np.array(img).flatten() for img in train['image']])
y_train = np.array(train['label'])
X_test = np.array([np.array(img).flatten() for img in test['image']])
y_test = np.array(test['label'])
num_classes = 10
models = []
for c in range(num_classes):
  y_bin = (y_train == c).astype(int)
  model = LogisticRegression(lr = 0.01, n_iters = 1000)
  model.fit(X_train, y_bin)
  models.append(model)
def predict_multiclass(X):
  probs = []
  for model in models:
    probs.append(model.predict_proba(X))
  probs = np.array(probs).T
  return np.argmax(probs, axis = 1)
preds = predict_multiclass(X_test)
accuracy = np.mean(preds == y_test)
print("Accuracy", accuracy)
```
